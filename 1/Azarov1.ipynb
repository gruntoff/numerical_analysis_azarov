```python
from tabulate import tabulate
from math import fabs

a = [2.1, 0.4, 1.3]
b = [-3.4, -3.7, 1.4]
c = [-2.7, 1.5, 2.5]
d = [7.2, 9.5, 5.8]

precision = 1e-5

class Esystem():
    
#     x1=(a1+b1x2+c1x3)|d1
#     x2=(a2+b2x1+c2x3)|d2
#     x3=(a3-b3x1-c3x2)|d3

    def __init__(self, a, b, c, d):
        v = a
        self.b = b
        self.c = c
        self.d = d
        
        self.iteration = int(1);
        self.x = [0, 0, 0]
        self.xTemp = [0, 0, 0]
        self.e = [0, 0, 0]
        self.maxE = 1
        
        self.listAll = []
        
    def get_maxE(self):
        return self.maxE
    
        
    def solve(self):
        
        self.xTemp[0] = (self.a[0] + (self.b[0] * self.x[1]) + (self.c[0] * self.x[2])) / self.d[0]
        self.xTemp[1] = (self.a[1] + (self.b[1] * self.x[0]) + (self.c[1] * self.x[2])) / self.d[1]
        self.xTemp[2] = (self.a[2] + (self.b[2] * self.x[0]) + (self.c[2] * self.x[1])) / self.d[2]
        
        self.e[0] = fabs(self.x[0] - self.xTemp[0])
        self.e[1] = fabs(self.x[1] - self.xTemp[1])
        self.e[2] = fabs(self.x[2] - self.xTemp[2])
        
        self.maxE = max(self.e[0], self.e[1], self.e[2])
        
        if (self.maxE > precision):
            self.append()
            self.iteration = self.iteration + 1
            self.x = self.xTemp.copy()
        else:
            self.printAll()
            self.printAnswer()
            
        
    def append(self):
        self.listAll.append([self.iteration] + self.xTemp + self.e + [self.maxE])
        
    def printAll(self):
        print(tabulate(self.listAll, headers=["Iteration", "x1", "x2", "x3", "e1", "e2", "e3", "e max"], floatfmt=".6f"))

    
    def printAnswer(self):
        print('\n Answer: x1 = %.5f, x2 = %.5f, x3 = %.5f' % (self.x[0], self.x[1], self.x[2]))
        

system = Esystem(a, b, c, d)
while True:
    system.solve()
    if system.get_maxE() < precision:
        break       
```
